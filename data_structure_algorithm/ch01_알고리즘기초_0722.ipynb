# ● 01 - 1 알고리즘이란?
    - 어떤 문제를 해결하기 위해 정해 놓은 일련의 절차
---
#### [ 구조 ]
    - 순차구조(sequential structure) : 한 문장씩 순서대로 처리되는 구조
    - 선택구조(select structure) : 조건식 평가결과에 따라 실행 흐름이 변경되는 구조

#### [ 조건문 ]
    - header ':' 앞에서 동작하는 코드 (if, while 등)
    - suite  ':' 뒤에서 실행하는 코드
    - 분기(branching) : 프로그램 실행 흐름을 다른 곳으로 변경하는 명령

#### [ 연산자 ]
    - 연산자 : operator, 피연산자 : operand
    - 단항 연산자 : 피연산자1 (ex) -a
    - 이항 연산자 : 피연산자2 (ex) a < b
    - 삼항 연산자 : 피연산자3 (ex) a if b else c
    - 조건 연산자 if~ else 는 파이썬의 유일한 삼항 연산자

#### [ 순서도 ]
    - flowcahrt : 문제를 정의, 분석하고 해결하는 방법을 그림으로 표현
    - 기호 종류
        - 데이터 : 기억장치를 지정하지 않은 데이터 자체 
        - 처리 : 여러 종류의 처리 기능
        - 미리 정의한 처리 :서브루틴이나 모듈 등
        - 판단 : 스위치형 기능
        - 루프 범위 
            - 시작기호, 종료기호 쌍을 이룸 (같은 이름)
            - 변수명 : 초깃값, 증가값, 종료값 (ex) i:1,2,n -> 시작이나 종료 둘 중 한 곳에 표기해야함
        - 선 : 제어흐름
        - 단말 : 외부환경으로부터의 출입(프로그램의 종료와 시작)

# ● 01-2 반복하는 알고리즘
---

#### [ 반복 구조 ]
    - (repetition structure) : loop
    - while: 카운터용 변수 사용
    - for : 변수가 하나만 있을 때 유리.
    - range : 이터러블 객체를 생성
    - 이터러블 객체 : 반복할 수 있는 객체 (list, str, tuple)

#### [ 무한루프와 break문 ] 
    - 파이썬에서는 while True, break를 통해 의도적으로 무한반복할 수 있다. 
    - 사후판단 반복문 : do ~ while, repeat~until, 그러나 파이썬은 사후판단 반복문을 제공하지 않으므로 break 사용

#### [ 구조적 프로그래밍 ] 
    - 순차, 선택, 반복 세 종류의 제어흐름을 사용.

#### [ 파이썬 객체 ] 
    - 데이터, 함수, 클래스, 모듈, 패키지 등
        - 객체는 자료형(data type)을 가지며 메모리 공간을 차지함.
        - 모든 객체는 식별번호(identity)를 가진다.
        - but 파이썬의 변수는 값을 가지지 않는다.
        - 파이썬의 변수는 객체에 연결된 이름에 불과하다. 변수는 그 값을 참조할 뿐.
        - 파이썬의 변수는 값과 같은 식별번호를 가진다. 값이 바뀌면 식별번호도 바뀐 값의 식별번호로 바뀜.


```python
# *을 n개 출력하되 w개마다 줄바꿈하기1

n = int(input('*을 출력할 개수를 입력하세요:'))
w = int(input('몇 개 마다 줄 바꿈할 까요?'))

for _ in range(n//w):
    print('*'* w)
    
rest = n % w
if rest:
    print('*' * rest)      
    
```

```python
# 양수만 입력받기
print('1부터 n까지 정수의 합을 구합니다.')

while True:
    n = int(input('n값을 입력하세요.: '))
    if n >0:
        break

sum=0
for i in range(1,n+1):
    sum+=i
print(f'1부터 {n}까지 정수의 합은 {sum}입니다.')    
```

```python
# 직사각형 넓이로 변의 길이 구하기
# 넓이가 area인 직사각형에서 변의 길이 나열하기(가로, 세로 길이는 정수)

area = int(input('가로,세로 길이를 구할 직사각형 넓이를 입력하세요.:'))

for i in range(1, area+1):
    # if i*i > area: break  #  직사각형 넓이에 해당하는 두 변 중 짧은 변이 짧은 변 중 가장 긴 변을 넘어가면 break.
                            #  -> 1 x 32, 32 x 1 처럼 두변이 같은 길이를 한 번만 계산함. 
    if area % i : continue  # 약수를 구하는 코드
                            # 나머지가 있다면 조건문 다음 코드를 실행하지 않고 다시 for문으로 돌아감.    
    print(f'{i} x {area // i}')

```

```python
# 10~99 사이의 난수 n개 생성하기(13이 나오면 중단)

import random

n = int(input('난수의 개수를 입력하세요.:'))

for _ in range(n):
    r = random.randint(10,99)
    print(r ,end =" ")
        
    if r == 13 :
        print('\n프로그램을 종료합니다.')
        break                                       # break는 그 순간, 코드 실행 종료
else: print('\n난수 생성을 종료합니다.')        
```

```python
# 반복문 건너뛰기와 여러 범위 스캔하기
# 1~12까지 8을 건너뛰고 출력하기1

# 건너뛸 값을 알 경우, 조건문 쓰지 않는 것이 효율적임.
for i in list(range(1,8)) + list(range(9,13)):
    print(i, end =' ')
print()

# 범위가 넓어질수록 비효율적인 방식    
for i in range(1,13):
    if i == 8: continue
    print(i, end = ' ')
        
    
```

```python
# 비교 연산자를 연속으로 사용하는 방법과 드로르간의 법칙  (예제) 2자리 양수(10~99) 입력받기
print('2자리 양수를 입력하세요.')

while True:
    no = int(input('값을 입력하세요.:'))
    if not(no<10 or no>99): # if 10 <= no <=99 와 같다. 
        break
    
print(f'입력받은 양수는 {no}입니다.')
```

```python
# 다중 루프 (예제) 구구단 곱셈표 출력하기
print('-'* 30 )

for i in range(2,10):
    for j in range(1,10):
        print(f'{i*j:3}',end=' ') # :3은 세자리로 일정한 간격을 가지게 해줌.
    print() 

print('-' * 30)       
```

```python
# 다중 루프 (예제) 이등변 삼각형으로 * 출력하기

# 왼쪽이 직각인 이등변 삼각형
print('왼쪽 아래가 직각인 이등변 삼각형을 출력합니다.')

n = int(input('짧은 변의 길이를 입력하세요.: '))
for i in range(n):
    for j in range(i+1):
        print('*',end='')
    print()    

# 오른쪽이 직각인 이등변 삼각형1
print('오른쪽 아래가 직각인 이등변 삼각형을 출력합니다.')
for i in range(n):
    for _ in range(n-(i+1)):
        print(" ", end ="")
    for _ in range(i+1):
        print('*', end="")
    print()        

# 오른쪽이 직각인 이등변 삼각형2
for i in range(n):
    for _ in range(n-(i+1)):
        print(" ", end ="")
    print("*" * (i+1), end="")
    print()
```
