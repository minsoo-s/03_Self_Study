# 타겟 데이터 체크

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## [1] 데이터 로딩
---

```python
target = pd.read_csv("target.csv")

date = list(pd.to_datetime(target["DATE"]))
ind = list(target["IND"])
civil = list(target["CIVIL"])
```

## [2] 데이터 체크

```python
plt.plot(date, ind)
plt.show()
```

```python
plt.plot(date, civil)
plt.show()
```
